```python
import glob
data_path = r'..\\data\\NYT\\*'

files = glob.glob(data_path)
l = []
for file in files:
    with open(file, 'r', encoding='utf-8', errors='ignore') as f:
        data = json.load(f)
        docs = data['response']['docs']
        for doc in docs:
            if 'keywords' in doc:
                keywords = doc['keywords']
                for key in keywords:
                    if key['name'] == 'persons':
                        l.append(key['value'])
n_dict={}
for each_l in l:
    if each_l in n_dict:
        n_dict[each_l] +=1
    else:
        n_dict[each_l] = 1

top = [(k, n_dict[k]) for k in sorted(n_dict, key=n_dict.get, reverse=True)]
```

```python
#Collecting the headlines from the NYT and adding them to list
files = glob.glob(data_path)
li = []
for file in files:
    with open(file, 'r', encoding='utf-8', errors='ignore') as f:
        try:
            for key in json.load(f)["response"]["docs"]:
                li.append(key["headline"]["main"])
        except:
            continue
print("HEADLINES FROM NEW YORK TIMES")
for elem in li[:15]:
    print(elem)
```

```python
#Finding the trending person from NYT in specidfied given duration
key_l = []
val_l = []
for k, v in top[:5]:
    key_l.append(k)
    val_l.append(v)

import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use('ggplot') 
plt.style.use('seaborn-white')

labels = key_l
sizes = val_l
explode = (0.1, 0.0, 0, 0,0)
colors = ['red', 'gold', 'lightskyblue', 'lightcoral', 'blue']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, colors=colors, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
plt.title("Most trending person in NYT from Jan-2017 to March-2017 ", y=1.18)
plt.show()
```

```python
#Printing the top 10 words from the Headlines of NYT
import re
wrd=[]
for w in li[:10]:
    res = re.split('\s+',w)
    for x in res:
        if x.isalpha():
            wrd.append(x)
n_wrd={}
for each_l in l:
    if each_l in n_wrd:
        n_wrd[each_l] +=1
    else:
        n_wrd[each_l] = 1

top = [(k, n_wrd[k]) for k in sorted(n_wrd, key=n_wrd.get, reverse=True)]  
```

```python
print("The top 10 words from headlines are as follow:")
print(top[:10])
```

# Using 2nd API i.e Archive and storing data in data folder

```python
import requests
import os
import urllib.request
import pandas as pd
import simplejson as json
import time
from itertools import chain
import csv

out_format = 'json'
nyt_key='26632b20578d4cfda8f9fdd65b9ed9da'
pages=10
offset = 0
dicts = []
for i in range(0, pages):
    offset = i
    url='https://api.nytimes.com/svc/archive/v1/2014/2.json'+'?'+ 'api-key=' + nyt_key
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    dict_response = json.loads(response.read().decode('utf-8'))
    response = dict_response['response']
    docs = response['docs']
    dicts.append(docs)
    time.sleep(2)
```

```python
path=r'..\\data\\Archive_API\\data.csv'
import unicodecsv as csv
toCSV = (list(chain.from_iterable(dicts)))
keys = toCSV[0].keys()
with open(path, 'wb') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(toCSV)
```
