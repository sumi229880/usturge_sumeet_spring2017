# Question 1: Prove if Zipfâ€™s Law works in Natural Language or not. Law works in Natural Language or not. 

```python
import glob
import re
wrds = []
new=[]
freq_word = {}
path = "C:/nltk_data/corpora/gutenberg/*.txt"
for filename in glob.glob(path):
    with open(filename, 'r') as f:
        for line in f:
            w = line.split()
            for x in w:
                wd = re.sub(r'\W', "", x)
                if wd.isalpha():
                    wrds.append(wd)
for y in wrds:
    if y not in freq_word:
        freq_word[y] = 1
    else:
        freq_word[y] += 1

#Sorting by Rank using LAMBDA func.I have used sorted because its been quick       

#top_words = [sorted(freq_word.items(), key=lambda x: x[1], reverse=True)]
top_words = [(k, freq_word[k]) for k in sorted(freq_word, key=freq_word.get, reverse=True)]
rank = 1

rnk = []
freq = []
import csv
with open('test.csv', 'w+', newline='') as csvfile:    
    spamwriter = csv.writer(csvfile, delimiter=',')
    spamwriter.writerow(["RANK","WORD","FREQUENCY"])  #importing records into csv file
    for k, v in top_words:    
        spamwriter.writerow([rank,k,v])
        rnk.append(rank)
        freq.append(v)
        rank += 1

```

```python
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use('ggplot') 
plt.style.use('dark_background')
plt.loglog(rnk,freq)
plt.title("Zipfâ€™s Law")
plt.xlabel('log(Rank)')
plt.ylabel('log(Frequency)')
plt.show()
```

# Conclusion: Lookinng at the plot of log(R) vs Log(F), it can be said that zipf's law holds valid for the set of files that I have used and it can be seen that it has slope of -1 which states that zipf's law holds valid for this set of gutenberg txt files.
