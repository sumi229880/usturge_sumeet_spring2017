# Question 1:
# Total words count in austen-persuasion.txt
# Unique words count in austen-persuasion.txt
# Frequency of each word
# Top 10 words with most frequency

```python
from nltk.corpus import gutenberg 
gutenberg.fileids() #files in gutenberg
```

```python
total_words = gutenberg.words('austen-persuasion.txt')
print(len(total_words)) #number of words
```

```python
total_unique = set(total_words)
print(len(total_unique)) #unique number of words
```

```python
wordcount={} #declaring empty dictonary
for word in total_words:
    if word.isalpha():     #adding only alphabateic words
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
```

```python
for k,v in wordcount.items():
    print (k, v) #frequency of each word
```

```python
top_words = [(k, wordcount[k]) for k in sorted(wordcount, key=wordcount.get, reverse=True)]
for k, v in top_words[:10]:
    print((k, v)) #top 10 words with highest frequency
```

```python
from collections import Counter   #Using Counter
c=Counter(wordcount)
print (c.most_common()[:10]) #top 10 words with highest frequency
```
